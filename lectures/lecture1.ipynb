**Погрешность измерения** – это отклонение результатов измерения от истинного значения измеряемой величины.

**Абсолютная ошибка** – разность между измеряемым (приближенным) значением $x_{изм}$ и истинной величиной $x_{ист}$
$$ \Delta x = x_{изм} - x_{ист}. $$

В общем случае само измеренное значение можно записать как $x_i$, в качестве истинного значения принято указывать его приближенное значение в виде среднего арифметического $\overline{x}$, поскольку собственно истинное значение $\tilde{x}$ обычно остается неизвестным. Поэтому величину абсолютной ошибки принято записывать в виде выражения
$$ \Delta x = \overline{x} - x_i $$

**Относительная ошибка** – погрешность измерения, выраженная отношением абсолютной погрешности измерения к истинному значению (точнее, к его приближенному значению):
$$ \delta = \frac{\Delta x}{\overline{x}} $$

Относительная погрешность является безразмерной величиной либо измеряется в процентах.
Абсолютная ошибка характеризует погрешность метода, который был выбран для измерения. Относительная ошибка характеризует качество измерений. Точностью измерения называют величину, обратную относительной ошибке, т.е. $1/\delta$

По характеру проявления различают **систематические**, **случайные** и **грубые** погрешности измерения.

1. Систематические ошибки порождается причинами, действующими регулярно и в определенном направлении. Они могут быть связаны с ошибками приборов (неправильная шкала, калибровка и т.п.), неучтенными экспериментатором; с индивидуальными ошибками экспериментатора; ошибками метода анализа и т.д.
Исключаются путем введения поправок, найденных экспериментальным путем (например, градуировка термопар).
2. Случайные ошибки являются составляющими погрешности измерения, изменяющиеся случайным образом в серии повторных измерений одной и той же величины, проведенных в одних и тех же условиях. В появлении таких ошибок не наблюдается какой-либо закономерности, они обнаруживаются при повторных измерениях одной и той же величины в виде некоторого разброса получаемых результатов. Случайные ошибки неизбежны, неустранимы и всегда присутствуют в результате измерения, однако их влияние, как правило, можно устранить статистической обработкой.
3. Грубые ошибки – погрешности, возникшие вследствие недосмотра экспериментатора или неисправности измерительной аппаратуры (например, неправильно зафиксированный номер деления на шкале прибора и т.п.).

В зависимости от характеристик измеряемой величины для определения
ошибок измерений используют различные методы:
– метод Корнфельда заключается в выборе интервала в пределах от
минимального xmin до максимального xmax результата измерений и
погрешность рассчитывается как половина разности между этими крайними
величинами измерения:
$$ \Delta x = \frac{x_{max} - x_{min}}{2} $$

– средняя квадратичная погрешность (среднеквадратичное отклонение)
Sn :
$$ S_n = \sqrt{\frac{\displaystyle\sum_{i=1}^n (x_i - \overline{x})^2}{n - 1}} $$
