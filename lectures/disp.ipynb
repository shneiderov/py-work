## Элементы дисперсионного анализа

Рассмотрим действие единичного фактора $А$ (количественного или качественного), который принимает $k$ различных значений (уровней фактора). На $i$-м уровне производится $n_i$
наблюдений, результаты которых можно записать следующим образом:

$$ \begin{bmatrix}
x_{1,1} & x_{2,1} & ... & x_{k,1} \\
x_{1,2} & x_{2,2} & ... & x_{k,2} \\
... & ... & ... & ... \\
x_{1,n_1} & x_{2,n_2} & ... & x_{k,n_k} \\
\end{bmatrix}. $$

Будем полагать, что результат любого наблюдения можно представить в виде модели:
$$ x_{i,j} = \mu + \beta_{i,j} + \varepsilon_{i,j}, $$

где $x_{i, j}$ – результат измерения параметра (отклик), полученный при $i$-м уровне фактора с $j$-м порядковым номером; \
    $\mu$ – точное значение $i$-го параметра (общее среднее); \
    $\beta_j$ – эффект (систематическая ошибка), обусловленный влиянием $i$-го уровня фактора выборки $x_{[n_j]}^{j}$; \
    $\varepsilon_{i, j}$ – случайная ошибка (вызванная влиянием неконтролируемых факторов).
