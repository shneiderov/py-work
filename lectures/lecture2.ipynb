## Основные обозначения

$X, Y$ – случайные величины, подлежащие исследованию.

Если $X$ – такая случайная величина,то её называют **генеральной совокупностью**. Для исследования производят $n$ независимых наблюдений, в которых получают значения $\overrightarrow{x} = (x_1, x_2, ..., x_n)$ – *выборку объёма* $n$ из генеральной совокупности. Если элементы выборки упорядочить $\big( x_{(1)}, x_{(2)}, ..., x_{(n)} \big)$, где $x_{(1)} \leqslant x_{(2)} \leqslant ... \leqslant x_{(n)}$, то результат называют **вариационным рядом**. Число $x_{(n)} - x_{(1)}$ называется **размахом выборки**.

Число $\displaystyle \overline{x} = \frac{1}{n} \sum_{i = 1}^n x_i$ называется **выборочным средним**.

Число $\displaystyle S_x^2 = S^2 = Dx = \frac{1}{n} \sum_{i = 1}^n \big( x_i - \overline{x} \big)^2 = \frac{1}{n}\sum_{i = 1}^nx_i^2-\overline{x^2}$ называется **выборочной дисперсией**. Число $S = \sigma x = \sqrt{Dx}$ – **выборочным средним квадратическим (стандартным) отклонением**.


## Гипотезы

**Статистической гипотезой** называется всякое высказывание о генеральной совокупности, проверяемое по результатам наблюдений (по выборке), не располагая сведениями о всей генеральной совокупности, по определенным правилам. По выборочным данным делают вывод о том, можно принять эту гипотезу или нет. Процедура сопоставления высказанной гипотезы с выборочными данными называется **проверкой гипотезы**. Обычно гипотезы обозначаются $H, H_0, H_1, ...$.

На первом шаге, имея выборочные данные и руководствуясь конкретными условиями рассматриваемой задачи, формулируют **нулевую** (основную) гипотезу $H_0$ и **альтернативную** гипотезу $H_1$, конкурирующую с гипотезой $H_0$. По данной выборке будет принято решение – справедлива ли гипотеза $H_0$, или гипотеза $H_1$. Необходимо отметить, что как правило не делается категорический вывод – "гипотеза $H_0$ истинна" или "гипотеза $H_0$ ложна". О гипотезе делается вывод – "вероятность справедливости гипотезы $H_0$ большая, чтобы ее принять", или "гипотеза $H_0$ не отвергается", или "гипотеза $H_0$ не противоречит выборочным данным".

На втором шаге задастся вероятность $\alpha$, которую называют **уровнем значимости**. Так как высказывание, что $H_0$ справедлива для генеральной совокупности принимается по выборочным данным, т.е. по ограниченному числу наблюдений, то решение может быть ошибочным. Если гипотеза $H_0$ отвергается (т.е., принимается гипотеза $H_1$), а на самом деле гипотеза $H_0$ верна, то совершается **ошибка первого рода**. Уровень значимости – это вероятность ошибки первого рода, равна $\alpha$. Вероятность $\alpha$ задается заранее малым числом. Обычно используются значения $\alpha$ : 0,05; 0,01; 0,005; 0,001. Для этих значений составлены таблицы.

Если принимается гипотеза $H_0$, а верна конкурирующая гипотеза $H_1$, то совершается **ошибка второго рода**. Вероятность сделать ошибку второго рода обозначается часто $\beta$. Отметим, что в результате проверки относительно гипотезы $H_0$ может быть принято и правильное решение. Вероятность этого решения равна $1 - \alpha$. Если не принимают гипотезу $H_0$ (т.е. принимают гипотезу $H_1$), тогда как и на самом деле гипотеза $H_0$ не верна, то вероятность этого решения равна $1 - \beta$.

На третьем шаге находят величину $\varphi = \varphi(x_1, ..., x_n)$, зависящую от случайной выборки $(x_1, ..., x_n)$. По ее значениям выносят решение о "расхождении выборки с гипотезой $H_0$". Величина $\varphi(x_1, ..., x_n)$ – случайная величина т.к. выборка $(x_1, ..., x_n)$ случайна и она подчиняется при выполнении гипотезы $H_0$ некоторому известному закону распределения, затабулированному, т.е. для которого составлены таблицы. Величину называют **критерием**.

На четвертом шаге выделяется подобласть значений $W$ критерия $\varphi$, которые свидетельствуют о существовании расхождения выборки с гипотезой $H_0$ и, следовательно, о невозможности принять гипотезу $H_0$. Подобласть $W$ называется **критической областью**. Пусть критическая область выделена. Тогда, если вычисленные по выборке значения $\varphi$ попадают в критическую область, то гипотеза $H_0$ отвергается и принимается гипотеза $H_1$. Ясно, что такое решение может оказаться ошибочным, и на самом деле гипотеза $H_0$ может быть справедливой. То есть, ориентируясь на критическую область можно совершить ошибку первого рода, вероятность которой задана заранее и равна $\alpha$.

Тема 8. Проверка гипотезы однородности средних и дисперсий

Проверяемая гипотеза о равенстве математических ожиданий (об однородности математических ожиданий) случайных величин в общем случае имеет вид
$$ H_0:\mu_1=\mu_2=...=\mu_m, $$
при конкурирующей
$$ H_1: \mu_{i_1} \ne \mu_{i_2}. $$

Для проверки гипотезы $H_0$ может использоваться ряд параметрических критериев: сравнения двух выборочных средних при известных дисперсиях; сравнения двух выборочных средних при неизвестных, но равных дисперсиях (критерий Стьюдента); сравнения двух выборочных средних при неизвестных и неравных дисперсиях; $F$-критерий. В этих же целях применяется целая совокупность непараметрических критериев: критерий Уилкоксона, критерий Манна-Уитни, критерий Краскела-Уаллиса.

Основным предположением, обусловливающим применение параметрических критериев, является **принадлежность анализируемых выборок нормальному закону**. Непараметрические критерии свободны от этого требования.
Проверяемая гипотеза о постоянстве дисперсии $m$ выборок имеет вид
$$ H_0:\sigma_1^2=\sigma_2^2=...=\sigma_m^2, $$
при конкурирующей
$$ H_1: \sigma_{i_1}^2 \ne \sigma_{i_2}^2, $$
где неравенство выполняется, по крайней мере, для одной пары индексов $i_1, i_2$. Для проверки такого вида гипотез применяется ряд критериев: критерий Бартлетта, критерий Кокрена, критерий Фишера, критерий Левене.

## Воспоминания

Оценка математического ожидания выборки $x$
$$ \overline{x} = \frac{1}{n-1}\sum_{i=1}^{n}x_i. $$

Сумма квадратов отклонений (*SS - sum of squares*) от математического ожидания выборки $x$
$$ SS_x = \sum_{i=1}^n\big(x_i-\overline{x}\big)^2. $$

Оценка дисперсии (среднеквадратического (стандартного) отклонения) выборки $x$
$$ S_x = SD_x^2 = \frac{1}{df}\sum_{i=1}^n\big(x_i-\overline{x}\big)^2. $$

При $x \sim N$
$$ df = n - 1. $$

Оценка стандартной ошибки выбоки $x$
$$ SE_x^2 = \frac{SD_x^2}{n}. $$

## t-test

Наибольшей популярностью при проверке гипотез о равенстве двух **генеральных средних** пользуется $t$-критерий Стьюдента.

$$H_0: \mu_x = \mu_y, x, y \sim N \quad vs \quad H_1: \mu_1 \ne \mu_2 \big| x, y \nsim N$$

Пусть $\big(x_1, x_2, ..., x_{n_x}\big)$, $\big(y_1, y_2, ..., y_{n_y}\big)$ – две выборки взаимно независимых случайных величин. По этим наблюдениям можно вычислить оценки $\displaystyle\overline{x} = \frac{1}{n_x}\sum_{i=1}^{n_x}x_i$ и $\displaystyle\overline{y} = \frac{1}{n_y}\sum_{i=1}^{n_y}y_i$ для неизвестных математических ожиданий $\mu_x$ и $\mu_y$, а также оценки $\displaystyle SD_x^2 = \frac{1}{n_x-1}\sum_{i=1}^{n_x}\big(x_i-\overline{x}\big)^2$ и $\displaystyle SD_y^2 = \frac{1}{n_y-1}\sum_{i=1}^{n_y}\big(y_i-\overline{y}\big)^2$ для неизвестных дисперсий $\sigma_x^2$ и $\sigma_y^2$.

Разность оценок $ \Delta = \overline{x} - \overline{y} $ при выполнении $H_0$ стремится к 0, соответственно $ SE_{\Delta}^2 = SE_x^2 + SE_y^2 = \cfrac{SD_x^2}{n_x} + \cfrac{SD_y^2}{n_y}$. Следовательно $t$-статистика для проверки нулевой гипотезы равна

$$ t = \cfrac{\Delta}{SE_{\Delta}} = \cfrac{\overline{x}-\overline{y}}{\sqrt{ \cfrac{SD_x^2}{n_x} + \cfrac{SD_y^2}{n_y} }}. $$

Данная статистика (случайные величины распределены по нормальному закону) подчиняется $t$-распределению Стьюдента с $df$-степенями свободы:

$$ df = n_x + n_y - 2.$$

При равенстве дисперсий $\sigma^2 = \sigma_x^2 = \sigma_y^2$ $t$-статистика для проверки нулевой гипотезы с учётом степеней свободы равна

$$ t = \cfrac{\overline{x}-\overline{y}}{\sqrt{ \cfrac{\sigma^2}{n_x} + \cfrac{\sigma^2}{n_y} }} = \cfrac{\overline{x}-\overline{y}}{\sigma \sqrt{ \cfrac{1}{n_x} + \cfrac{1}{n_y} }} = \cfrac{\overline{x}-\overline{y}}{\sqrt{\cfrac{(n_x - 1)SD_x^2 + (n_y - 1)SD_y^2}{n_x+n_y-2}} \cdot \sqrt{ \cfrac{1}{n_x} + \cfrac{1}{n_y} }}.$$

При условии сравнения не двух выборок, а одной выборки с генеральной совокупностью

$$ t = \cfrac{\Delta}{SE_{\Delta}} = \cfrac{\overline{x}-\mu_x}{\sqrt{ \cfrac{SD_x^2}{n_x}}}. $$


## Пример 1

```python
import scipy.stats as stats
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sb

matplotlib.style.use('ggplot')
```

```python
np.random.seed(6)

rvs = stats.norm.rvs(scale=5, size=200000, loc=45)
pop_ages, minsk_ages = np.split(rvs, [150000])

fig, ax = plt.subplots()
ax.hist(pop_ages, bins=20, edgecolor='w', density=True, color='r', alpha=0.5, label='Ген сов')
ax.hist(minsk_ages, bins=20, edgecolor='w', density=True, color='g', alpha=0.5, label='Выборка')
ax.set_xlabel('Возраст, год')
ax.set_title('Нормализованное гистограммы распределения')
ax.legend()
plt.show()
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))
for ax in axs:
    ax.set_ylabel('Возраст, год')
axs[0].violinplot([pop_ages, minsk_ages], showmeans=True, points=20, widths=[0.75, 0.75])
axs[0].set_xticks([1, 2], labels=['Ген сов', 'Выборка'])
axs[1].boxplot([pop_ages, minsk_ages], tick_labels=['Ген сов', 'Выборка'], widths=[0.5, 0.5])
plt.show()
```

```python
alpha = 0.05
m = pop_ages.mean()
mx = minsk_ages.mean()
sd = pop_ages.std(ddof=1)
sdx = minsk_ages.std(ddof=1)
se = sd / np.sqrt(len(pop_ages))
sex = sdx / np.sqrt(len(minsk_ages))
df = len(minsk_ages) - 1

t = (mx - m) / sex
t
```

```python
fig, ax = plt.subplots()
x = np.linspace(stats.t.ppf(0.01, df), stats.t.ppf(0.99, df), 100)
ax.plot(x, stats.t.pdf(x, df), 'r-', lw=1)
ax.plot(x, stats.t.pdf(x, 1), 'r--', lw=1, alpha=0.5)
ax.set_ylim(bottom=0, top=0.45)
ax.vlines([t], 0, 0.45, color='b', lw=1, alpha=0.5)
ax.hlines([0.05], -2.3, 2.3, color='b', lw=1, alpha=0.5)
ax.fill_between(x, stats.t.pdf(x, df), np.zeros(len(x)), alpha=0.2)
ax.set_title(r'Форма распределения Стьюдента для степени свободы df=' + str(df), fontsize=12)
plt.show()
```

```python
if stats.t.pdf(t, df) < alpha:
    print("Отклоняем гипотезу H0. Существует различие между средними выборки и генеральной совокупности.")
else:
    print("Принимаем гипотезу H0. Не существует статистических различий между средними наборов данных.")
```

```python
stats.ttest_1samp(minsk_ages, popmean=pop_ages.mean())
```

```python
t, p = stats.ttest_1samp(minsk_ages, popmean=pop_ages.mean())
if p < alpha:
    print("Отклоняем гипотезу H0. Существует различие между средними выборки и генеральной совокупности.")
else:
    print("Принимаем гипотезу H0. Не существует статистических различий между средними наборов данных.")
```

```python
alpha = 0.05
mx = setosa['sepal_length'].mean()
my = versicolor['sepal_length'].mean()
sdx = setosa['sepal_length'].std(ddof=1)
sdy = versicolor['sepal_length'].std(ddof=1)
sex = sdx / np.sqrt(setosa['sepal_length'].count())
sey = sdy / np.sqrt(versicolor['sepal_length'].count())
df = setosa['sepal_length'].count() + versicolor['sepal_length'].count() - 2

t = (mx - my) / np.sqrt(sex**2 + sey**2)
t
```

```python
t, p = stats.ttest_ind(setosa['sepal_length'], versicolor['sepal_length'])
t
```

```python
if stats.t.pdf(t, df) < alpha:
    print("Отклоняем гипотезу H0. Существует различие между средними выборки и генеральной совокупности.")
else:
    print("Принимаем гипотезу H0. Не существует статистических различий между средними наборов данных.")
```

# Дисперсионный анализ

## Введение

Дисперсионный анализ (от латинского *Dispersio* – рассеивание) – статистический метод выявления различий между выборочными средними для трёх и более выборок.
Целью дисперсионного анализа является проверка значимости различия между средними с помощью сравнения дисперсий. Дисперсию измеряемого признака разлагают на независимые слагаемые, каждое из которых характеризует влияние того или иного фактора или их взаимодействия. Последующее сравнение таких слагаемых позволяет оценить значимость каждого
изучаемого фактора, а также их комбинации.

Однофакторный (одномерный) дисперсионный анализ (One-Way ANOVA – ANalysis Of VAriance) – метод дисперсионного анализа, в котором рассматривается только один фактор (независимая переменная) и две или более группы.





Для применения дисперсионного анализа независимые переменные должны быть категориальными (неметрическими). Они называются ФАКТОРАМИ.

Многофакторный (многомерный) дисперсионный анализ -- метод дисперсионного анализа, в котором рассматривается много факторов (независимых переменных).

Многомерный дисперсионный анализ (MANOVA) -- это процедура сравнения многомерных выборочных средних.

Как многомерная процедура, она используется при наличии двух или более зависимых переменных и часто сопровождается тестами значимости с участием отдельных зависимых переменных по отдельности.

## Однофакторный дисперсионный анализ (One-Way ANOVA)


Имеется $k \ge 3$ **независимых** выборок

$$x_{[n_1]}^{1}, ..., x_{[n_k]}^{k},$$

которые являются **количественными** наблюдениями непрерывных случайных величин
$$ x_{i, j} = \mu_i + \beta_{i, j} + \varepsilon_{i, j}, \qquad i = 1, ..., n, \qquad j = 1, ..., k,$$

$$ x_{i, j} = \mu + \beta_j + \varepsilon_{i, j}, \qquad i = 1, ..., n_j, \qquad j = 1, ..., k,$$
где $x_{i, j}$ – результат измерения исследуемой величины, полученной при $i$-м уровне фактора с $j$-м порядковым номером;
    $\mu$ – точное значение $i$-го параметра (общее среднее);
    $\beta_j$ – эффект (систематическая ошибка), обусловленный влиянием $i$-го уровня фактора выборки $x_{[n_j]}^{j}$;
    $\varepsilon_{i, j}$ – случайная ошибка (вызванная влиянием неконтролируемых факторов).

Основные предпосылки дисперсионного анализа:
- нормальное распределение случайных величин
- равенство дисперсий
- математическое ожидание возмущения $\varepsilon_{i, j}$ равно нулю для любых $i$, т.е. 

Предположим $\mu_j = \mu + \beta_j$.
- $H_0: \mu_1 = \mu_2 = ... = \mu_k$,
- $H_1: \exist i,j: \mu_i \ne \mu_j$
